Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Preprocessing

```python
data_trace = pd.read_csv('csv_example\Delta_R1.csv', sep=',', header=19)
data_trace_motors = pd.read_csv('csv_example\M31.csv', sep=',', header=8)

data_trace_1 = data_trace.drop(['Index', 'Date', 'ClockTime', 'RawTime', 'TraceSampID'], axis = 'columns')
data_trace_motors_1 = data_trace_motors.drop(['Index', 'Offset'], axis = 'columns')
data_trace_motors_1 = data_trace_motors_1.loc[0:1023]

for index in range(len(data_trace_1.columns)):
    data_trace_1[data_trace_1.columns[index]] = data_trace_1[data_trace_1.columns[index]].replace(',','.',regex=True)
    data_trace_1[data_trace_1.columns[index]] = pd.to_numeric(data_trace_1[data_trace_1.columns[index]], downcast="float") 

for index in range(len(data_trace_motors_1.columns)):
    data_trace_motors_1[data_trace_motors_1.columns[index]] = pd.to_numeric(data_trace_motors_1[data_trace_motors_1.columns[index]], downcast="float") 
```

Plot

```python
data_trace_2 = data_trace_1.loc[5000:5500]

plt.figure(figsize=(20, 6))
plt.title("DR1_A1")

plt.plot(data_trace_2[data_trace_1.columns[0]], label=data_trace_1.columns[0])
plt.plot(data_trace_2[data_trace_1.columns[1]], label=data_trace_1.columns[1])
plt.plot(data_trace_2[data_trace_1.columns[2]], label=data_trace_1.columns[2])
plt.legend(loc="upper left")
plt.show() 

plt.figure(figsize=(20, 6))
plt.title("DR1_A2")

plt.plot(data_trace_2[data_trace_1.columns[12]], label=data_trace_1.columns[12])
plt.plot(data_trace_2[data_trace_1.columns[13]], label=data_trace_1.columns[13])
plt.plot(data_trace_2[data_trace_1.columns[14]], label=data_trace_1.columns[14])
plt.legend(loc="upper left")
plt.show() 


plt.figure(figsize=(20, 6))
plt.title("DR1_A3")

plt.plot(data_trace_motors_1[data_trace_motors_1.columns[0]], label=data_trace_motors_1.columns[0])
#plt.plot(data_trace_motors_1[data_trace_motors_1.columns[1]], label=data_trace_motors_1.columns[1])
plt.plot(data_trace_motors_1[data_trace_motors_1.columns[2]], label=data_trace_motors_1.columns[2])
plt.plot(data_trace_motors_1[data_trace_motors_1.columns[3]], label=data_trace_motors_1.columns[3])
plt.legend(loc="upper left")
plt.show() 
```
