Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Preprocessing

```python
data_trace = pd.read_csv('Trace1.csv', sep=',', header=19)

data_trace['Axis0.Cmd.Pos:LREAL'] = data_trace['Axis0.Cmd.Pos:LREAL'].replace(',','.',regex=True)
data_trace['Axis0.Cmd.Pos:LREAL'] = pd.to_numeric(data_trace['Axis0.Cmd.Pos:LREAL'], downcast="float") 

data_trace['Axis0.Cmd.Vel:LREAL'] = data_trace['Axis0.Cmd.Vel:LREAL'].replace(',','.',regex=True)
data_trace['Axis0.Cmd.Vel:LREAL'] = pd.to_numeric(data_trace['Axis0.Cmd.Vel:LREAL'], downcast="float")

data_trace['Axis0.Act.Pos:LREAL'] = data_trace['Axis0.Act.Pos:LREAL'].replace(',','.',regex=True)
data_trace['Axis0.Act.Pos:LREAL'] = pd.to_numeric(data_trace['Axis0.Act.Pos:LREAL'], downcast="float")

data_trace['Axis0.Cmd.Trq:LREAL'] = data_trace['Axis0.Cmd.Trq:LREAL'].replace(',','.',regex=True)
data_trace['Axis0.Cmd.Trq:LREAL'] = pd.to_numeric(data_trace['Axis0.Cmd.Trq:LREAL'], downcast="float")

data_trace_1 = data_trace.drop(['Date', 'ClockTime', 'RawTime', 'TraceSampID'], axis = 'columns')
```

Plot

```python
plt.figure(figsize=(20, 8))
plt.title("Sysmac Data Trace")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.plot(data_trace_1['Index'], data_trace_1['Axis0.Cmd.Pos:LREAL'])
plt.plot(data_trace_1['Index'], data_trace_1['Axis0.Cmd.Vel:LREAL'])
plt.plot(data_trace_1['Index'], data_trace_1['Axis0.Act.Pos:LREAL'])
plt.plot(data_trace_1['Index'], data_trace_1['Axis0.Cmd.Trq:LREAL'])
plt.show() 
```

```python
data_trace_1[['Axis0.Cmd.Pos:LREAL', 'Axis0.Cmd.Vel:LREAL', 'Axis0.Act.Pos:LREAL', 'Axis0.Cmd.Trq:LREAL']] = data_trace_1[['Axis0.Cmd.Pos:LREAL', 'Axis0.Cmd.Vel:LREAL', 'Axis0.Act.Pos:LREAL', 'Axis0.Cmd.Trq:LREAL']].apply(lambda x: x / (x.max() - x.min()))

data_trace_2 = data_trace_1.loc[0000:10000]

plt.figure(figsize=(20, 8))
plt.title("Sysmac Data Trace")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.plot(data_trace_2['Index'], data_trace_2['Axis0.Cmd.Pos:LREAL'])
plt.plot(data_trace_2['Index'], data_trace_2['Axis0.Cmd.Vel:LREAL'])
plt.plot(data_trace_2['Index'], data_trace_2['Axis0.Act.Pos:LREAL'])
plt.plot(data_trace_2['Index'], data_trace_2['Axis0.Cmd.Trq:LREAL'])
plt.show()
```
